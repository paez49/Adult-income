# Analysis the null values

```python
def count_null(df):
    """Count nulls fields in the dataframe."""
    null_per_col = df.isna().sum()
    return null_per_col
```

```python
count_null(df)
```

No null data is found, however, some rows with "?" are found, which would be the same thing.

```python
def count_question_marks(df):
    """
    This function takes a df and a symbol representing null values,
    and returns the count of null values in each column.
    """
    df = df.replace("?", None)
    nulls_per_column = df.isna().sum()
    
    return nulls_per_column
```

```python
count_question_marks(df)
```

# Categorical Variables Encoding

## Using label encoder

```python
label_encoder = LabelEncoder()

df['race'] = label_encoder.fit_transform(df['race'])
df['gender'] = label_encoder.fit_transform(df['gender'])
df["income"] = label_encoder.fit_transform(df['income'])

df
```

## Using one hot encoder 

```python
df=pd.get_dummies(df, columns=['workclass'])
df=pd.get_dummies(df, columns=['native-country'])
df=pd.get_dummies(df, columns=['occupation'])
df=pd.get_dummies(df, columns=['marital-status'])
df
```

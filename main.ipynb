```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("adult.csv")
df
```

```python
df.count()
```

```python
df_with_nan = df.dropna()
df_with_nan
```

### Delete useless index

```python
df_2_delete = df[df["workclass"] == "?"]
df_2_delete

bad_index = df_2_delete.index

df = df.drop(bad_index)
df
```

### Delete null values

```python
df = df.dropna()
df
```

```python
ages = df['age']
plt.hist(ages, bins=20, edgecolor='k')  # Ajustar el número de bins según sus preferencias
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades')
plt.show()
```

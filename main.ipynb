# Data cleaning

### Delete useless index

```python
df_2_delete = df[(df["workclass"] == "?") & (df["occupation"] == "?")]
bad_index = df_2_delete.index

df = df.drop(bad_index)
rows_with_question_mark = df[df.eq("?").all(axis=1)]
rows_with_question_mark
```

### Delete null values

```python
df = df.dropna()
df
```
